# KNN with Iris


## Libraries

```python
import sklearn.neighbors as nei
import pandas as pd
import sklearn.model_selection as mod
```

## Data

```python
# Load the iris data set from a URL.
df = pd.read_csv('Iris.csv')
```

```python
df
```

## Visualise

```python
# Load the seaborn package
import seaborn as sns

# Plot the Iris data set with a pair plot
sns.pairplot(df, hue ='class')
```

## Inputs and Outputs

```python
inputs = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
outputs = df['class']
```

## Classifier

```python
knn = nei.KNeighborsClassifier(n_neighbors=5)
```

## Fit

```python
knn.fit(inputs, outputs)
```

## Predict

```python
df.loc[121]
```

```python
knn.predict([[5.6,2.8,4.9,2.0], [4.0,2.0,1.0,4.0]])
```

## Evaluate

```python
(knn.predict(inputs) == outputs).sum()  #returning a numpy array. This will convert all Trues to 1 and Falses to 0 so counting up we can see if there are any False values
```

```python
inputs_train, inputs_test, outputs_train, outputs_test = mod.train_test_split(inputs,outputs,test_size=0.33)
```

```python
knn = nei.KNeighborsClassifier(n_neighbors=5)
knn.fit(inputs_train, outputs_train)
```

```python
(knn.predict(inputs_test) == outputs_test).sum()
```
